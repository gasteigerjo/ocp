# Get validation data sample

```python
from ocpmodels.common.registry import registry

val_dataset_config = {'src': 'data/s2ef/all/val_id/'}

val_dataset = registry.get_dataset_class('trajectory_lmdb')(val_dataset_config)

len(val_dataset)
```

```python
import numpy as np

np.random.seed(123)
validation_sample_indices = np.random.choice(range(len(val_dataset)), size=30000)

len(validation_sample_indices)
```

```python
validation_sample = []
for i, validation_sample_index in enumerate(validation_sample_indices):
    validation_sample.append(val_dataset[validation_sample_index])
    print(i)
```

```python
len(validation_sample)
```

```python
validation_sample[0]
```

```python
import pickle
import lmdb
from tqdm import tqdm

db_path = 'val_sample.lmdb'

db = lmdb.open(
    db_path,
    map_size=1099511627776 * 2,
    subdir=False,
    meminit=False,
    map_async=True,
)

for data_object in tqdm(validation_sample, total=len(validation_sample)):
    txn = db.begin(write=True)
    txn.put(
        f"{data_object.fid}".encode("ascii"),
        pickle.dumps(data_object, protocol=-1),
    )
    txn.commit()

# Save count of objects in lmdb.
txn = db.begin(write=True)
txn.put("length".encode("ascii"), pickle.dumps(len(validation_sample), protocol=-1))
txn.commit()

db.sync()
db.close()
```
